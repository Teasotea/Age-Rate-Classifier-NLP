# Scrapping IMDB


```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
#url = 'https://www.imdb.com/list/ls000634294/'
pages = np.arange(1, 9)
movie_name =[]
movie_year = []
movie_age_rate = []
movie_runtime = []
movie_genre = []
movie_descr = []
```

```python
for page in pages:
    page = requests.get("https://www.imdb.com/list/ls000634294/?sort=list_order,asc&st_dt=&mode=detail&page="+str(page), headers = {"Accept-Language": "en-US, en;q=0.5"})
    soup = BeautifulSoup(page.text, 'html.parser')
    movie_data = soup.findAll('div', attrs = {'class': 'lister-item mode-detail'})
    #print(movie_data)
    sleep(randint(1, 8))
    
    for store in movie_data:
        age_rate = store.p.find('span', class_ = 'certificate')
        if (age_rate!=None):
            movie_age_rate.append(age_rate.text)
            print(age_rate.text)
        else: 
            continue
        
        name = store.h3.a.text
        movie_name.append(name)
        sleep(randint(2, 8))
        
        year = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace(')', '').replace('(', '').replace("I ","").replace(" Video", "")
        movie_year.append(year)
        
        runtime = store.p.find('span', class_ = 'runtime').text.replace(' min', '')
        movie_runtime.append(runtime)
    
        genre = store.p.find('span', class_ = 'genre').text.strip('\n').strip('\t')
        movie_genre.append(genre)
    
        descr= store.find('p', class_ = '').text.strip('\n').strip('\t')
        movie_descr.append(descr)
        
```

```python
print(np.count_nonzero(movie_name))
print(np.count_nonzero(movie_year))
print(np.count_nonzero(movie_age_rate))
print(np.count_nonzero(movie_runtime))
print(np.count_nonzero(movie_genre))
print(np.count_nonzero(movie_descr))
```

```python
movie_df  = pd.DataFrame({'Name of movie': movie_name, 'Description': movie_descr, 'Age rate': movie_age_rate, 'Year': movie_year, 'Runtime': movie_runtime, 'Genre': movie_genre})
```
